```python
import numpy as np

from activations import Sigmoid
from loss import CrossEntropyLoss


X = np.array([[1,-2,0.4,2],
             [2,4,-5,1],
             [5,6,-1,0.4]])
y = np.array([0,1,0])
w = np.array([0.2,0.1,0.5,0.9])
b = np.ones(shape=(1,))
lr = 0.1

for i in range(20):
    # b = np.array([1,])
    sigm = Sigmoid()
    loss_cls = CrossEntropyLoss()
    pred = sigm(X@w.T + b)
    print("Predict = ",pred)
    loss = loss_cls.loss(y, pred)
    print("Loss = ",loss)
    d_w, d_b = loss_cls.decent_loss(X,y,w,b)
    print("D_w = ",d_w)
    print("D_b = ",d_b)
    d_w = sigm.decent(d_w)
    d_b = sigm.decent(d_b)
    print("W = ",w)
    print("B = ",b)
    w -= d_w * lr
    b -= d_b * lr
```

```python
from sklearn.datasets import make_classification

from classifiers import LogReg
from loss import CrossEntropyLoss
from optimizers import FullGD


X, y = make_classification(random_state=42,n_classes=2,n_features=5,n_samples=10000)

optimizer = FullGD(loss=CrossEntropyLoss())
model = LogReg(optimizer=optimizer,lr=0.01)
# model.predict_proba(X)
model.fit(X,y,max_iter=100,test_size=0.2,verbose="plot")

```
